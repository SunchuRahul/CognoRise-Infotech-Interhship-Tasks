# Exploratory Data Analysis-EDA

```python
sns.barplot(x=df["calories"],y=df["protein"])
plt.title("Calories vs Protein")
plt.show()
```

```python
pd.crosstab(df["fat"],["sugars"]).plot(kind="line", stacked=True)
plt.title("According to sugars how much fat Increased")
plt.show()
```

```python
plt.pie(df["vitamins"].value_counts(),autopct="%2f%%",labels=df["vitamins"].unique())
plt.title("vitamins")
plt.show()
```

```python
plt.hist(df["fiber"], color="red")
plt.title("Fiber")
plt.show()
```

```python
plt.scatter(df["mfr"],df["name"])
plt.title("Relationship")
plt.ylabel("mfr")
plt.xlabel("name")
plt.show()
```

```python
sns.heatmap(df[["carbo","fiber"]],fmt=".1f")
```

```python
sns.heatmap(df[["name","mfr","calories","protein","fat","sodium","fiber","carbo","sugars","potass","vitamins"]].corr(),annot=True)
plt.show()
```
