```python
plt.figure(figsize=[10,5])
sns.heatmap(d.corr(),annot=True)
```

```python
sns.pairplot(d, vars=['alcohol', 'citric acid', 'density', 'residual sugar'], hue='quality')
plt.suptitle('Pair Plot of Selected Features by Wine Quality', y=1.02)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='quality', y='alcohol', data=d)
plt.title('Box Plot of Alcohol Content by Wine Quality')
plt.show()
```

```python
plt.hist(d['quality'], bins=6, edgecolor='black')
plt.xlabel('Wine Quality')
plt.ylabel('Frequency')
plt.title('Distribution of Wine Quality')
plt.show()
```

```python
sns.pairplot(d, vars=['alcohol', 'pH', 'residual sugar'], hue='quality', markers='o')
plt.suptitle('Pair Plot of Selected Features by Wine Quality', y=1.02)
plt.show()
```
